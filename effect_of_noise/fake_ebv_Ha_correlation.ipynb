```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('science')
plt.rcParams.update({'font.size': 15})
```

```python
N = 1000
Ha = np.random.uniform(1, 2, size=N)
ratio = np.random.uniform(2.86, 4.0, size=N)
Hb = Ha / ratio

Ha_obs = Ha + np.random.normal(0, 1, size=N) * np.abs(Ha) / 5
Hb_obs = Hb + np.random.normal(0, 1, size=N) * np.abs(Hb) / 5
```

```python
plt.figure(figsize=(4, 4), dpi=400)
plt.scatter(Ha, Hb/Ha)
plt.xlabel(r'$H\alpha$')
plt.ylabel(r'$H\beta/H\alpha$')
```

```python
plt.figure(figsize=(4, 4), dpi=400)
plt.scatter(Ha_obs, Hb_obs / Ha_obs)
plt.xlabel(r'$H\alpha$')
plt.ylabel(r'$H\beta/H\alpha$')
```

```python
N = 1000
Ha = np.random.uniform(1, 2, size=N)
ratio = np.random.uniform(2.86 + (Ha - 1.5) / 2, 4.0 + (Ha - 1.5) / 2, size=N)
Hb = Ha / ratio

Ha_obs = Ha + np.random.normal(0, 1, size=N) * np.abs(Ha) / 5
Hb_obs = Hb + np.random.normal(0, 1, size=N) * np.abs(Hb) / 5
```

```python
plt.figure(figsize=(4, 4), dpi=400)
plt.scatter(Ha, Hb/Ha)
plt.xlabel(r'$H\alpha$')
plt.ylabel(r'$H\beta/H\alpha$')
```

```python
plt.figure(figsize=(4, 4), dpi=400)
plt.scatter(Ha_obs, Hb_obs / Ha_obs)
plt.xlabel(r'$H\alpha$')
plt.ylabel(r'$H\beta/H\alpha$')
```
